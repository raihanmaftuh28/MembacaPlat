```python
from PIL import Image
import pytesseract
import cv2
import os
import matplotlib.pyplot as plt

image = cv2.imread("plat.jpg")
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
color = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
(thresh, blackAndWhite) = cv2.threshold(gray, 127, 0, cv2.THRESH_BINARY)
plt.imshow(color)
plt.show()
```

```python
white_so = (200, 200, 200)
white_real = (255, 255, 255)
mask = cv2.inRange(color,white_so,white_real)
ready = cv2.bitwise_and(color,color,mask=mask)
plt.imshow(ready)
plt.show()
```

```python
filename = "{}.png".format(os.getpid())
cv2.imwrite(filename,ready)
```

```python
text = pytesseract.image_to_string(Image.open(filename))
os.remove(filename)
print(text)
```
